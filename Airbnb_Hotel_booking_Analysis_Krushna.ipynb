### Function to calculate Lower and Upper Outlier Bounds(IQR Method)

```python
def remove_outlier(col_name):
    sorted(col_name)
    Q1,Q3 = col_name.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return lower_bound, upper_bound
```

### Data Cleaning, Conversion, and CalculatingOutlier Bounds for Service fee

```python
# Convert 'price' and 'service fee' to numeric, handling currency symbols and commas
df['price'] = df['price'].astype(str).str.replace('$', '').str.replace(',', '').astype(float)
df['service fee'] = df['service fee'].astype(str).str.replace('$', '').str.replace(',', '').astype(float)

low, high = remove_outlier(df['service fee'])
```

```python
df['service fee'] = np.where(df['service fee']>high,high,df['service fee'])
```

```python
df['service fee'] = np.where(df['service fee']<low,low,df['service fee'])
```

```python
sns.boxplot(y='service fee',data=df)
```

### Calculating Outlier Bounds for minimum nights

```python
low, high = remove_outlier(df['minimum nights'])
```

```python
df['minimum nights'] = np.where(df['minimum nights']>high,high,df['minimum nights'])
```

```python
df['minimum nights'] = np.where(df['minimum nights']<low,low,df['minimum nights'])
```

```python
sns.boxplot(y='minimum nights',data=df)
```

### Distribution of Price, Service Fee, and Minimum Nights by Neighbourhood Group

```python
numerical_cols = ['price', 'service fee', 'minimum nights']

plt.figure(figsize=(15, 5 * len(numerical_cols)))

for i, col in enumerate(numerical_cols):
    plt.subplot(len(numerical_cols), 1, i + 1)
    sns.boxplot(x="neighbourhood group", y=col, data=df, palette="Set2")
    plt.title(f"Box Plot of {col.replace('_', ' ').title()} by Neighbourhood Group")
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```

### Note: The box plots show that Manhattan has the highest median price, with a wide distribution. Service fees and minimum nights also vary significantly across the neighbourhood groups, with several outliers observed in some areas.
